```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("covid_vaccine_statewise.csv")
```

```python
data.head(5)
```

```python
data.tail(5)
```

```python
data.describe()
```

```python
data.dtypes
```

```python
data["State"].unique()
data = data[data['State'] != 'India']
```

```python
data.isnull().sum()
```

```python
data['First Dose Administered'].replace(0 , np.nan , inplace = True)
data['First Dose Administered'].fillna(data['First Dose Administered'].mean() , inplace = True)
data['Second Dose Administered'].replace(0 , np.nan , inplace = True)
data['Second Dose Administered'].fillna(data['Second Dose Administered'].mean() , inplace = True)
data['Male (Doses Administered)'].replace(0 , np.nan , inplace = True)
data['Male (Doses Administered)'].fillna(data['Male (Doses Administered)'].mean() , inplace = True)
data['Female (Doses Administered)'].replace(0 , np.nan , inplace = True)
data['Female (Doses Administered)'].fillna(data['Female (Doses Administered)'].mean() , inplace = True)
data['Total Doses Administered'].replace(0 , np.nan , inplace = True)
data['Total Doses Administered'].fillna(data['Total Doses Administered'].mean() , inplace  = True)
data.fillna(0 , inplace=True)
data.head()
```

```python
data = data.drop(['Transgender (Doses Administered)' ,'18-44 Years (Doses Administered)' , '45-60 Years (Doses Administered)' , '60+ Years (Doses Administered)' , '18-44 Years(Individuals Vaccinated)' , '45-60 Years(Individuals Vaccinated)' , '60+ Years(Individuals Vaccinated)' , 'Transgender(Individuals Vaccinated)' , 'Sessions'] , axis = 1)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data['Total Doses Administered'] = data['Total Doses Administered'].astype('int64')
data['First Dose Administered'] = data['First Dose Administered'].astype('int64')
data['Second Dose Administered'] = data['Second Dose Administered'].astype('int64')
data['Male (Doses Administered)'] = data['Male (Doses Administered)'].astype('int64')
data['Female (Doses Administered)'] = data['Female (Doses Administered)'].astype('int64')
data[' Covaxin (Doses Administered)'] = data[' Covaxin (Doses Administered)'].astype('int64')
data['CoviShield (Doses Administered)'] = data['CoviShield (Doses Administered)'].astype('int64')
data['Sputnik V (Doses Administered)'] = data['Sputnik V (Doses Administered)'].astype('int64')
data['AEFI'] = data['AEFI'].astype('int64')
data['Male(Individuals Vaccinated)'] = data['Male(Individuals Vaccinated)'].astype('int64')
data['Female(Individuals Vaccinated)'] = data['Female(Individuals Vaccinated)'].astype('int64')
data['Total Individuals Vaccinated'] = data['Total Individuals Vaccinated'].astype('int64')
```

```python
data.dtypes
```

```python
date = data['Updated On'].str.split('/', expand=True)
date
```

```python
data['day'] = date[0]
data['month'] = date[1]
data['year'] = date[2]

data['day'] = pd.to_numeric(data['day'])
data['month'] = pd.to_numeric(data['month'])
data['year'] = pd.to_numeric(data['year'])

data['Updated On'] = pd.to_datetime(data['Updated On'])
data.rename(columns = {'Updated On' : 'Vaccine_date'}, inplace = True)

data.head()
```

```python
data.describe()
```

```python
number_of_males_vaccinated = data['Male (Doses Administered)'].sum()
males_to_plot = number_of_males_vaccinated
number_of_males_vaccinated = '{:,.0f}'.format(number_of_males_vaccinated)
print('Total Number of Males Vaccinated : ',number_of_males_vaccinated )
```

```python
number_of_females_vaccinated = data['Female (Doses Administered)'].sum()
females_to_plot = number_of_females_vaccinated
number_of_females_vaccinated = '{:,.0f}'.format(number_of_females_vaccinated)
print('Total Number of Females Vaccinated : ',number_of_females_vaccinated ) 
```

```python
plt.pie([males_to_plot , females_to_plot], labels = ['Males','Females'],
        colors = ['#4f46e5','#ec4899'], autopct = '%0.0f%%', startangle = 90, radius = 2)
plt.legend(loc='best',shadow=True,fontsize=11)
plt.show()
```

```python
new_data = data
```

```python
statewise_vaccination = new_data.groupby('State')['Total Individuals Vaccinated'].sum().to_frame('Total Individuals Vaccinated')
statewise_vaccination_to_plot = statewise_vaccination.copy()
statewise_vaccination['Total Individuals Vaccinated'] = statewise_vaccination['Total Individuals Vaccinated'].apply(lambda x: '{:,.0f}'.format(x))
print(statewise_vaccination)
```

```python
statewise_vaccination['Total Individuals Vaccinated'] = statewise_vaccination['Total Individuals Vaccinated'].str.replace(',', '').astype(int)

statewise_vaccination_to_plot.plot(kind='bar', figsize=(17,10))
plt.ylabel('Number of people vaccinated')
plt.show()
```
